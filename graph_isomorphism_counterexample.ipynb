
## Proof by Counterexample

Let \( A = (V_1, E_1) \) and \( B = (V_2, E_2) \), such that

\[
V_1 = \{a, b, c\}, \quad E_1 = \{(a, b), (b, c)\}
\]

\[
V_2 = \{x, y, z\}, \quad E_2 = \{(x, y), (y, z)\}
\]

Now observe that \( |V_1| = 3 \) and \( |V_2| = 3 \). However, the degree \( \deg(u) \) of any \( u \in V_1 \cup V_2 \) is in the set \( \{1, 2\} \).

This shows us that neither graph is completely connected, because a completely connected graph (i.e., a complete graph) has \( \deg(u) = n - 1 \) for all \( u \in V \), where \( V \) is the vertex set of a complete graph of order \( n \).

With that in mind, we can now construct an isomorphism.

Let \( \phi \) be a function \( \phi: V_1 \to V_2 \) defined by the mapping:

\[
\phi:
\begin{pmatrix}
a & b & c \\
x & y & z
\end{pmatrix}
\]

This is obviously a bijective function. Now we just need to verify that the edges map properly.

- \( (a, b) \in E_1 \), and \( \phi((a, b)) = (x, y) \in E_2 \)
- \( (b, c) \in E_1 \), and \( \phi((b, c)) = (y, z) \in E_2 \)

Therefore, \( A \cong B \), even though neither \( A \) nor \( B \) is a complete graph.

We have thus shown that it **cannot** be the case that for two graphs \( A', B' \), \( A' \cong B' \) **if and only if** \( A' \) and \( B' \) are completely connected.
